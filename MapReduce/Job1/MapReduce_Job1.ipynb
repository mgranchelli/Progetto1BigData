```python
import re
# as per recommendation from @freylis, compile once only
CLEANR = re.compile('<.*?>') 

def cleanhtml(raw_html): 
    return re.sub(CLEANR, ' ', raw_html)
```

```python
import sys
import string
with open('./reviews_job1_dim_05.csv') as f:
    contents = f.readlines()
    #print(contents)

    
i = 0

# read line from standard input
for line in contents:
    # removing leading/trailing whitespaces
    line = line.strip()
    # split the current line into words
    year, words = line.split("\t")
    
    # remove html tag inside text
    words = cleanhtml(words)
    # replace dot with withespace
    words = words.replace(".", " ")
    
    # remove punctuation
    words = words.translate(str.maketrans('', '', string.punctuation))
    words = re.sub(' +', ' ', words)
    words = words.strip()
    # splittare riga in words per il campo text
    words = words.split(" ")
    
    for word in words:
        # write in standard output the mapping word -> 1
        # in the form of tab-separated pairs
        with open('./test_output.csv', 'a') as w:
            line = year + '\t' + word.lower() + '\t' + str(1) + '\n'
            w.writelines(line)
        #print('%s\t%s\t%i' % (year, word.lower(), 1))
w.close()
f.close()
```

```python
with open('./test_output.txt') as f:
    contents = f.readlines()
year_words = {}

# input comes from STDIN
# note: this is the output from the mapper
for line in contents:

    # as usual, remove leading/trailing spaces
    line = line.strip()

    # parse the input elements
    year, current_word, current_count = line.split("\t")

    # convert count (currently a string) to int
    try:
        current_count = int(current_count)
        year = int(year)
    except ValueError:
        # count was not a number, in this case is just int
        continue

    # initialize word that were not seen befor with 0
    if year not in year_words:
        year_words[year] = {}
    
    if current_word not in year_words[year]:
        year_words[year][current_word] = 0
    
    year_words[year][current_word] += current_count

for year, words in sorted(year_words.items(), reverse=True):
    print('%i' % year)
    for word in sorted(words.items(), key=lambda x: x[1], reverse = True)[:10]:
        print('\t%s\t%i' % (word[0], word[1]))
```
