```python

import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt

def plot_job(job_data, job_n):
    plt.figure(figsize=(10,5))
    fig = sns.lineplot(data=job_data, dashes=False, marker="o", linewidth = 3)
    plt.xlabel("Dataset Size")
    plt.ylabel("Time (Second)")
    plt.title("Execution Times - Job"+str(job_n))
    plt.show(fig)
```

```python
plot_job(job1, 1)
```

```python
plot_job(job2, 2)
```

```python
plot_job(job3, 3)
```
